### Text Lemmatization

Lemmatization usually refers to doing things properly with the use of a vocabulary and morphological analysis of words, normally aiming to remove inflectional endings only and to return the base or dictionary form of a word, which is known as the lemma.

For grammatical reasons, documents are going to use different forms of a word, such as organize, organizes, and organizing. Additionally, there are families of derivationally related words with similar meanings, such as democracy, democratic, and democratization. 

These different words are used as per condition but all of them have same meaning and have same importance and use in case of sentiment analysis. In Text Classification, we need to represent each text word in the form of number so in case of organize, organizes, and organizing we will be needing to use these 3 same words and represent it wil different numbers that will increase the dimension of our dataset. 

The Lemmatization helps there and it transforms any word to it base word. For example democracy, democratic, and democratization will be transformed into democracy every where in the text and we will only need 1 dimension to represnt these multiple words. 

```python
# Lemmatize the text
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()
```

```python
def lemmatize_text(text):
    return ' '.join([lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)])

df['clean_review'] = df['clean_review'].apply(lemmatize_text)
# Make lemmatized tokens to sentence
# df['lemmatized'] = df['lemmatized'].str.join(' ')
```

```python
df.head()
```

From Data Frame above mentioned, we analyzed that there are many single characters left in data, which are the remaings of the data cleaning.
These words probably dont add any value in the data. So, we will delete the characters from the text.

But, first lets Analyze these words!

```python
# For words with len 1
single_char = []
count= 0
for i in df.clean_review:
    for j in i.split():
        if len(j)==1 and j.isdigit()==False:
            count+=1
            single_char.append(j)

print('There are total single chracter words remaing in the data: ',count )
```

```python
# Lets Remove these Characters.
# Single characters could also be digits and we make sure that the char should be alphabets.
def remove_single_chars(text):
    clean = []
    for i in text.split():
        if len(i)==1 and i.isalpha()==True:
            pass
        else:
            clean.append(i)
    return ' '.join(clean)

    
df['clean_review'] = df['clean_review'].apply(remove_single_chars)
```

#### Label Encoding

```python
le = LabelEncoder()
le.fit(df['sentiment'])
df['encoded_sentiments'] = le.transform(df['sentiment'])
```

#### Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(df['clean_review'],
                                                    df['encoded_sentiments'],
                                                    test_size = 0.25,
                                                    random_state = 7)
```

```python
print('Shape of X_train', X_train.shape)
print('Shape of X_test', X_test.shape)
print('Shape of Y_train', Y_train.shape)
print('Shape of Y_test', Y_test.shape)
```

#### Feature Transformation

```python
from sklearn.feature_extraction.text import TfidfVectorizer

max_feature_num = 50000
vectorizer = TfidfVectorizer(max_features=max_feature_num)
train_vecs = vectorizer.fit_transform(X_train)

test_vecs = TfidfVectorizer(max_features=max_feature_num,
                            vocabulary=vectorizer.vocabulary_).fit_transform(X_test)
```

```python
train_vecs.shape, test_vecs.shape
```
